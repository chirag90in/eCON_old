```python
import nilearn as nil
from nilearn import plotting
```

```python
shock = nil.image.load_img('/data/bswift-1/Pessoa_Lab/eCON/dataset/Group/ShockUncensored/UnconVsCon_paired/splitted_regs/noProx/shock_contrast.nii.gz')
STATEmean = nil.image.load_img('/data/bswift-1/Pessoa_Lab/eCON/dataset/Group/ShockUncensored/UnconVsCon_paired/splitted_regs/noProx/STATEmean_contrast.nii.gz')
STATEdiff = nil.image.load_img('/data/bswift-1/Pessoa_Lab/eCON/dataset/Group/ShockUncensored/UnconVsCon_paired/splitted_regs/noProx/STATEdiff_contrast.nii.gz')
TRAITmean = nil.image.load_img('/data/bswift-1/Pessoa_Lab/eCON/dataset/Group/ShockUncensored/UnconVsCon_paired/splitted_regs/noProx/TRAITmean_contrast.nii.gz')
TRAITdiff = nil.image.load_img('/data/bswift-1/Pessoa_Lab/eCON/dataset/Group/ShockUncensored/UnconVsCon_paired/splitted_regs/noProx/TRAITdiff_contrast.nii.gz')
```

## Shock (uncon - con)

```python
threshold = 3.4729
```

```python
plotting.view_img(shock,threshold=threshold,title='Shock (uncon-con)',dim=0,vmax=10,draw_cross=False)
```

Consistent with previous results. Same regions show up, except this time there is a bigger cluster in the PCC for **con > uncon**.

## STATE mean

```python
plotting.view_img(STATEmean,threshold=threshold,title='STATE mean',dim=0,vmax=10,draw_cross=False)
```

## TRAIT mean

```python
plotting.view_img(TRAITmean,threshold=threshold,title='TRAIT mean',dim=0,vmax=10,draw_cross=False)
```

## STATE difference

```python
plotting.view_img(STATEdiff,threshold=threshold,title='STATE difference',dim=0,vmax=10,draw_cross=False)
```

## TRAIT difference

```python
plotting.view_img(TRAITdiff,threshold=threshold,title='TRAIT difference',dim=0,vmax=10,draw_cross=False)
```
