```python
import pandas as pd
import numpy as np
import os
from os.path import join, isdir, dirname, basename, exists
from os import mkdir

import nibabel as nib

from scipy.stats import t as tdist
import glob as glob
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(context="paper",style="white",font_scale=1.5)
```

```python
yoked = pd.read_excel('../../onsetdir/CON_yoked_table.xlsx')
yoked = yoked.query('use == 1').reset_index()
```

```python
appr_window_TRs = list(range(2,8))
shock_window_TRs = list(range(8,15))

appr_area = ['PCC']
shock_area = ['Right BNST','Left BNST']
```

```python
main_df = pd.read_csv('PerSubjShockResp.csv')
main_df.head()
```

```python
def getMeanResp(main_df):
    df_list = []
    for cond, window in zip(['shock','appr'],[shock_window_TRs,appr_window_TRs]):
        df = main_df[(main_df['TR'].isin(window))]
        df = df.groupby(['Subject','ROI','Group','Pair'])['Response'].mean().reset_index()
        df.rename(columns={'Response':cond},inplace=True)
        df_list.append(df)
    return pd.merge(df_list[0],df_list[1],on=[col for col in df_list[0].columns if col not in ['shock','appr']],how='left')
```

```python
new_df = getMeanResp(main_df)
new_df = new_df.melt(id_vars=['Subject','ROI','Group','Pair'],value_vars=['shock','appr'],var_name='cond',value_name='beta')
```

```python
new_df.head()
```

```python
data = new_df[new_df['ROI'].isin(appr_area+shock_area)].reset_index(drop=True)
data.head()
```

```python
data = data.pivot_table(values='beta',index=['Subject','Group','Pair','cond'],columns=['ROI']).reset_index()
data.head()
```

```python
from scipy import stats
def r2(x, y):
    return stats.pearsonr(x, y)[0] ** 2

pal = sns.color_palette()
mypal = [pal[2],pal[0]]
```

```python
fig, axes = plt.subplots(ncols=2,figsize=(15,6))
axes= axes.flatten()
for i, roi in enumerate(['Right BNST','Left BNST']):
    con_appr = data[(data['cond']=='appr') & (data['Group']=='control')]['PCC'].values
    con_shock = data[(data['cond']=='shock') & (data['Group']=='control')][roi].values
    
    con_corr= stats.pearsonr(con_appr,con_shock)
    
    uncon_appr = data[(data['cond']=='appr') & (data['Group']=='uncontrol')]['PCC'].values
    uncon_shock = data[(data['cond']=='shock') & (data['Group']=='uncontrol')][roi].values
    
    uncon_corr= stats.pearsonr(uncon_appr,uncon_shock)
    
    sns.regplot(x=con_appr,y=con_shock,ax=axes[i],color=mypal[0],label='Control (r=%.2f, p=%.2f)' %(con_corr))
    sns.regplot(x=uncon_appr,y=uncon_shock,ax=axes[i],color=mypal[1],label='Uncontrol (r=%.2f, p=%.2f)' %(uncon_corr))
    axes[i].legend()
    axes[i].set_xlabel('PCC')
    axes[i].set_ylabel(roi)
    
fig.suptitle('Approach to shock analysis',fontsize=16)
```

```python
con_shock_resp_df = main_df[(main_df['TR'].isin(shock_window_TRs)) & (main_df['ROI'].isin(shock_area)) & (main_df['Group']=='control')]
con_shock_resp_df = con_shock_resp_df.groupby(['Subject','ROI'])['Response'].mean().unstack()

con_appr_resp_df = main_df[(main_df['TR'].isin(appr_window_TRs)) & (main_df['ROI'].isin(appr_area)) & (main_df['Group']=='control')]
con_appr_resp_df = con_appr_resp_df.groupby(['Subject','ROI'])['Response'].mean().unstack()

control = con_shock_resp_df.join(con_appr_resp_df)
```
