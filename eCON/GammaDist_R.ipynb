```r
alpha = 2
beta = 10
avrg = alpha*beta
std.dv = sqrt(alpha*beta^2)
x = 3
# Draw the gamma distribution
range = seq(0, avrg+5*std.dv, 0.01)
y_pdf = dgamma(range,alpha,rate = 1/beta)
y_cdf = pgamma(range,alpha,rate = 1/beta)
plot(range, y_pdf, type='l',ylim=c(0,max(y_pdf)+0.01))
# Add area to the left of x (<3)
cord.a = c(0,seq(min(range),x,0.01),x)
cord.b = c(0,dgamma(seq(min(range),x,0.01),alpha,rate = 1/beta),0)
polygon(cord.a,cord.b,col = 'blue')


plot(range, y_cdf, type='l',ylim=c(0,max(y_cdf)+0.01))
# Add area to the left of x (<3)
cord.a = c(0,seq(min(range),x,0.01),x)
cord.b = c(0,pgamma(seq(min(range),x,0.01),alpha,rate = 1/beta),0)
polygon(cord.a,cord.b,col = 'blue')
```
