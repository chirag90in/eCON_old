```python
import pandas as pd
import numpy as np
import os
from os.path import join, isdir, dirname, basename, exists
from os import mkdir
#import nibabel as nib
import glob as glob
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(context="paper",style="white",font_scale=1.5)
```

```python
bucket_path = '../../dataset/results_ShockCensored/{subj}/{group}lable/splitted_regs/ROI_{roi}/{subj}_bucket_REML_clean.1D'
yoked = pd.read_excel('../../onsetdir/CON_yoked_table.xlsx')
yoked = yoked.query('use == 1')
```

```python
effects = [effect+cat for effect in 'prox sign mag signXmag'.split() for cat in 'Nearest Near Far'.split()]
nuisance = 'v_motion onset_block offset_block'.split()
contrasts = 'prox_Nearest_v_Far prox_Nearest_v_Near prox_Near_v_Far prox sign_Nearest_v_Far sign_Nearest_v_Near sign_Near_v_Far sign mag_Nearest_v_Far mag_Nearest_v_Near mag_Near_v_Far mag signXmag_Nearest_v_Far signXmag_Nearest_v_Near signXmag_Near_v_Far signXmag'.split()
ALLEFFECTS = effects + nuisance + contrasts
colnames=['Pair','Subject','Group','ROI']
```

```python
rois = 'vmPFC PCC1 PCC2 rvAI rdAI rPAG1 lPAG1 PAG2 PAG3 rHippocampus lHippocampus rHabenula lHabenula rBNST  lBNST rAmygdala1 lAmygdala1 rAmygdala2 rAmygdala3 rAmygdala4 A1 MT'.split()
```

```python
if exists('Individual_betas_from_ROI_analysis_SC.csv'):
    main_df = pd.read_csv('Individual_betas_from_ROI_analysis_SC.csv')
else:
    main_df = pd.DataFrame()
    for i, row in yoked.iterrows():
        pair = i
        for kind in ['control','uncontrol']:
            for roi in rois:
                data = np.loadtxt(bucket_path.format(subj=row[kind],group=kind,roi=roi))[1::2]
                tmp_df = pd.DataFrame(data,index=ALLEFFECTS).T
                tmp_df['Pair'] = i
                tmp_df['Subject'] = row[kind]
                tmp_df['Group'] = kind
                tmp_df['ROI'] = roi
                main_df = pd.concat([main_df,tmp_df],axis=0)
    main_df.to_csv('Individual_betas_from_ROI_analysis_SC.csv',index=False)
```

```python
def paired_pivot(main_df, ValueVarList):
    df = pd.DataFrame()
    for valueVar in ValueVarList:
        for roi in main_df['ROI'].unique():
            tmp_df = main_df[['Pair','Group',valueVar]][main_df['ROI']==roi].pivot(index='Pair',columns='Group',values=valueVar)
            tmp_df.rename(index=str, columns={"control": "control_{effect}_{roi}".format(effect=valueVar,roi=roi),
                                              "uncontrol": "uncontrol_{effect}_{roi}".format(effect=valueVar,roi=roi)},inplace=True)
            df = pd.concat([df,tmp_df],axis=1)
    return df
```

```python
df = paired_pivot(main_df,ALLEFFECTS)
```

```python
new_df = main_df[main_df['ROI'].isin(ROIS)]
new_df.head()
```

```python
# Looping thru each pair
id_cols = ['Pair','Group','Subject','ROI']
for contrast in contrasts:
    rba = pd.DataFrame()
    for i in new_df.Pair.unique():
        pair = new_df[id_cols+[contrast]][new_df['Pair']==i]
        diff = pair[contrast][pair['Group']=='uncontrol'].values - pair[contrast][pair['Group']=='control'].values
        tmp_df = pd.DataFrame(diff,index=ROIS,columns=['Y'])
        tmp_df.index.name = 'ROI'
        tmp_df['Subj'] = 'P%i' %(i)
        tmp_df.reset_index(inplace=True)
        rba = pd.concat([rba,tmp_df])
    rba[['Subj','Y','ROI']].to_csv('../../dataset/data_for_RBA/uncon_v_con_{}.txt'.format(contrast),index=False,sep='\t',float_format='%.4f')
```

```python
# Looping thru each pair
id_cols = ['Subject','ROI']
for contrast in contrasts:
    for group in new_df.Group.unique():
        tmp_df= new_df[id_cols +[contrast]][new_df['Group']==group]
        mapper = {"Subject":"Subj",contrast:"Y"}
        tmp_df.rename(columns=mapper,inplace=True)
        tmp_df[['Subj','Y','ROI']].to_csv('../../dataset/data_for_RBA/{group}_{contrast}.txt'.format(group=group,contrast=contrast),index=False,sep='\t',float_format='%.4f')
```

```python
def plotbox(main_df, effect, ROI):
    #pal = sns.color_palette("bright")
    pal = sns.color_palette()
    mypal = [pal[2],pal[0]]
    emapper = {'sign': 'Direction (Approach vs. Retreat)','prox':'Proximity','mag':'Speed','signXmag':'Direction X Speed'}
    #Title = emapper[effect.split('_')[0]]+ ': ' + effect.split('_')[-3] + ' vs. '+ effect.split('_')[-1]
    Title = emapper[effect.split('_')[0]]
    simpleEffectColname = [effect.split('_')[0]+effect.split('_')[i] for i in [-3,-1]]
    tmp_df = main_df[colnames + simpleEffectColname]
    tmp_df = tmp_df.melt(id_vars=colnames,value_vars=simpleEffectColname,var_name='effect',value_name='beta')
    tmp_df['Effect'] = tmp_df['effect'].apply(lambda x: x.replace(effect.split('_')[0],''))
    plt.figure(figsize=(10,8))
    ax = sns.violinplot(x='Group',y='beta',data=tmp_df[tmp_df['ROI']==ROI],
                        hue='Effect',split=False,saturation=1,palette=mypal,
                        inner='quartile',linewidth=2)
    #ax = plt.gca()
    plt.setp(ax.collections, alpha=0.75)
    ax.set_alpha(0.5)
    handles, labels = ax.get_legend_handles_labels(); 
    sns.stripplot(x='Group',y='beta',data=tmp_df[tmp_df['ROI']==ROI], dodge=True,
                  palette=mypal,hue="Effect",jitter=True,alpha = 0.7,size=4,
                  edgecolor='black',linewidth=1)
    #sns.boxenplot(x='Group', y='beta',data=tmp_df[tmp_df['ROI']==ROI], hue='Effect' ,palette=mypal, width=0.5)
    #sns.boxplot(x='Group',y='beta',data=tmp_df[tmp_df['ROI']==ROI],hue='Effect',saturation=0.75,palette=mypal,width=0.8,whis=1.5)
    #sns.swarmplot(x='Group',y='beta',data=tmp_df[tmp_df['ROI']==ROI], linewidth=1, dodge=True, palette=mypal,hue="Effect",split=True)
    
    plt.title(ROI+'\n'+Title,fontsize=18)
    plt.ylabel('Estimated Activtion (A.U.)',fontsize=18)
    plt.xlabel('')
    ax.legend(handles, labels,loc='center right',bbox_to_anchor=(1.05, 0.9),fontsize=16)
    #plt.legend(loc='center right',bbox_to_anchor=(1.05, 0.9))
    plt.axhline(y = 0,linestyle='--',linewidth='2',color='black')
    plt.tick_params(axis='x',labelsize=18)
    #plt.ylim(-0.06,0.06)
    plt.tight_layout()
    return ax
    #plt.savefig('figures/interaction_plots/{effect}_{ROI}_boxswarm.png'.format(effect=effect,ROI=ROI),dpi=600)
```
